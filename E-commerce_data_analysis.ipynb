2. Data Preprocessing and Cleaning

```python
df.dtypes
```

```python
# column type handling

df['item_id'] = df['item_id'].astype('Int64')
df['qty_ordered'] = df['qty_ordered'].astype('Int64')
df['created_at'] = pd.to_datetime(df['created_at'], format='mixed', dayfirst=True)
df['Working Date'] = pd.to_datetime(df['Working Date'], format='mixed', dayfirst=True)
df['Customer Since'] = pd.to_datetime(df['Customer Since'], format='mixed', dayfirst=True)
df['M-Y'] = pd.to_datetime(df['M-Y'], format='mixed', dayfirst=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(columns=['Unnamed: 0','FY','Working Date',' MV '],inplace=True)  # not needed these columns
```

```python
df.columns = df.columns.str.strip()
```

3. Data Filtering

```python
print(df['payment_method'].unique())
print(df['status'].unique())
print(df['category_name_1'].unique())
```

```python
print(df['status'].value_counts())  # Count of order statuses
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['Customer Since'] = pd.to_datetime(df['Customer Since'])
```

```python
# Total Sales (completed orders)
completed_orders = df[df['status'] == 'complete']
total_sales = completed_orders['grand_total'].sum()
print(f"Total Sales: {total_sales}")
```

```python
# Canceled Orders
canceled_orders = df[df['status'] == 'canceled']
print(f"Canceled Orders: {len(canceled_orders)}")
```

```python
# Most Ordered Categories
print(df['category_name_1'].value_counts().head(10))
```

4. Exploratory Data Analysis (EDA):

```python
# Add Month column to completed_orders as well
completed_orders['Month'] = completed_orders['created_at'].dt.to_period('M')

# Group by Month and calculate total sales
monthly_sales = completed_orders.groupby('Month')['grand_total'].sum()

# Plotting
monthly_sales.plot(kind='line', marker='o', figsize=(10, 5), title='Monthly Sales Trend')
plt.ylabel('Sales (PKR)')
plt.xlabel('Month')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Selling Products
top_products = completed_orders.groupby('sku')['qty_ordered'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(10, 5), title='Top 10 Selling Products')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming you already have sizes and labels
sizes = df['payment_method'].value_counts().values
labels_raw = df['payment_method'].value_counts().index

# Calculate percentages
percentages = [f"{label} - {size / sum(sizes) * 100:.1f}%" for label, size in zip(labels_raw, sizes)]

# Optional explode
explode = [0.05] * len(sizes)

# Plot
plt.figure(figsize=(8, 8))
patches, texts = plt.pie(
    sizes,
    explode=explode,
    startangle=140
)

# Add legend with percentages
plt.legend(patches, percentages, loc='best', bbox_to_anchor=(1, 0.5), fontsize=12)
plt.title('Payment Methods', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Discount Analysis
plt.figure(figsize=(10, 5))
plt.hist(df['discount_amount'], bins=30, edgecolor='black')
plt.title('Discount Amount Distribution')
plt.xlabel('Discount Amount')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
